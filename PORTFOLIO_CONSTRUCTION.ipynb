# Reading in Files

```python
stocks_df = pd.read_csv("stocks_fama_recession.csv", parse_dates = ["Unnamed: 0"])
stocks_df.set_index("Unnamed: 0", inplace=True)
stocks_df.index.name = "Date"
stocks_df.head()
```

# Further Data Cleaning

## Converting Prices of stocks to absolute values

```python
# Using column indexing to exclude last 5 columns
stocks_df.iloc[:, :-5] = stocks_df.iloc[:, :-5].abs()
```

## Isolating Sin Stocks Only

```python
# Removing the NaN value in sin_list
sin_list = sin_list[~pd.isnull(sin_list)]
sin_list
```

```python
sin_stocks = stocks_df[sin_list]
sin_stocks.head()
```

## Finding Percentage Change in Stocks

```python
returns_df = sin_stocks.pct_change()
returns_df
```

### Removing null values

```python
returns_df.dropna(how = "all", inplace=True)
returns_df
```

## Calculating Individual Weights

```python
total_no_sin = len(returns_df.columns)
total_no_sin
```

### Calculating Number of Active Stocks for Each Date
- Deducting number of NaN values per row from the total number of sin stocks to derive the number of active sin stocks for a given date

https://www.udemy.com/tutorial/finance-data-with-python-and-pandas/creating-the-equally-weighted-portfolio/

```python
active_stocks_num = total_no_sin - returns_df.isnull().sum(axis = 1)
active_stocks_num
```

```python
for (i, z), j in zip(returns_df.iterrows(), range(len(active_stocks_num))):
    # boolean_mask = pd.isnull(z)

    # mask = pd.notnull(z[1])
    # print(z.where(mask))
    # for x, y in boolean_mask:
    #     if y == False:
    #         print(z)

    for x in z.to_list():
        if pd.notnull(x):
            print(x)
```

```python
range(len(active_stocks_num))
```

```python
invest_list = []
returns_list = []

for i, k in zip(returns_df.iterrows(), range(len(active_stocks_num))):
    if pd.notnull(i):
        weight = 1/active_stocks_num.iloc[k]
        invest_list.append(i)
        returns_list.append(weight * j)
    else:
        continue
    print("Current Period: " + str(returns_df.index[k]))
    print("Stocks that we are currently investing in: " + ",".join(invest_list))
    print ("Total Return of Portfolio: " + str(sum(returns_list)))
    invest_list = []
```
